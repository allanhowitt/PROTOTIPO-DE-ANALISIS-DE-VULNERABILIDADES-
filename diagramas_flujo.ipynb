```python
# anexo 4
import graphviz

def crear_diagrama_flujo_tokenizacion():
    # Crear un nuevo gráfico dirigido
    dot = graphviz.Digraph('diagrama_flujo_tokenizacion', comment='Diagrama de flujo del proceso de tokenización')
    
    # Configurar el estilo global
    dot.attr('node', shape='box', style='filled', fillcolor='lightblue', fontname='Arial')
    dot.attr('edge', fontname='Arial')
    dot.attr(rankdir='TD', size='8,11', dpi='300')
    
    # Definir los nodos
    dot.node('A', 'Inicio', shape='oval')
    dot.node('B', 'Importar AutoTokenizer\nde transformers')
    dot.node('C', 'Inicializar tokenizador\npara t5-base')
    dot.node('D', 'Obtener ejemplo de\ndiverse_data[0]')
    dot.node('E', 'Extraer texto de entrada (func)\ny texto de etiqueta (cwe)')
    dot.node('F', 'Tokenizar texto de entrada\ncon max_length=512\ntruncation=True\npadding=max_length\nreturn_tensors=pt')
    dot.node('G', 'Tokenizar texto de etiqueta\ncon max_length=2\ntruncation=True\npadding=max_length\nreturn_tensors=pt')
    dot.node('H', 'Imprimir tokens de entrada\n(primeros 10)')
    dot.node('I', 'Imprimir longitud total\nde tokens de entrada')
    dot.node('J', 'Imprimir tokens\nde etiqueta')
    dot.node('K', 'Imprimir longitud\nde tokens de etiqueta')
    dot.node('L', 'Fin', shape='oval')
    
    # Definir las conexiones
    dot.edge('A', 'B')
    dot.edge('B', 'C')
    dot.edge('C', 'D')
    dot.edge('D', 'E')
    dot.edge('E', 'F')
    dot.edge('E', 'G')
    dot.edge('F', 'H')
    dot.edge('H', 'I')
    dot.edge('G', 'J')
    dot.edge('J', 'K')
    dot.edge('I', 'L')
    dot.edge('K', 'L')
    
    # Renderizar y guardar el gráfico
    try:
        dot.render('diagrama_flujo_tokenizacion_ANX_4', format='png', cleanup=True)
        print(f"Diagrama de flujo generado exitosamente como 'diagrama_flujo_tokenizacion.png'")
        return True
    except Exception as e:
        print(f"Error al generar el diagrama con graphviz: {e}")
        return False

# Alternativa usando matplotlib y networkx
def crear_diagrama_matplotlib():
    try:
        import matplotlib.pyplot as plt
        import networkx as nx
        
        # Crear grafo
        G = nx.DiGraph()
        
        # Añadir nodos
        nodos = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L"]
        etiquetas = {
            "A": "Inicio",
            "B": "Importar AutoTokenizer",
            "C": "Inicializar tokenizador t5-base",
            "D": "Obtener ejemplo de diverse_data[0]",
            "E": "Extraer texto entrada y etiqueta",
            "F": "Tokenizar texto entrada (max_length=512)",
            "G": "Tokenizar texto etiqueta (max_length=2)",
            "H": "Imprimir tokens entrada (10 primeros)",
            "I": "Imprimir longitud tokens entrada",
            "J": "Imprimir tokens etiqueta",
            "K": "Imprimir longitud tokens etiqueta",
            "L": "Fin"
        }
        G.add_nodes_from(nodos)
        
        # Añadir conexiones
        conexiones = [
            ("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), 
            ("E", "F"), ("E", "G"), ("F", "H"), ("G", "J"),
            ("H", "I"), ("J", "K"), ("I", "L"), ("K", "L")
        ]
        G.add_edges_from(conexiones)
        
        # Dibujar el grafo
        plt.figure(figsize=(10, 14))
        pos = nx.planar_layout(G)  # o nx.spring_layout(G) para un layout diferente
        nx.draw(G, pos, with_labels=True, labels=etiquetas, 
                node_color="lightblue", node_size=3000, font_size=8,
                arrows=True, arrowsize=15)
        
        # Guardar figura
        plt.savefig("diagrama_flujo_tokenizacion_matplotlib_ANX_4.png", dpi=300, bbox_inches="tight")
        print("Diagrama básico generado con matplotlib como 'diagrama_flujo_tokenizacion_matplotlib.png'")
        return True
    except ImportError:
        print("Se requieren las bibliotecas matplotlib y networkx. Puedes instalarlas con:")
        print("pip install matplotlib networkx")
        return False

# Podemos también generar un diagrama utilizando la sintaxis de Mermaid (para integrar con Markdown)
def generar_codigo_mermaid():
    mermaid_code = """
flowchart TD
    A[Inicio] --> B[Importar AutoTokenizer de transformers]
    B --> C[Inicializar tokenizador para t5-base]
    C --> D[Obtener ejemplo de diverse_data[0]]
    D --> E[Extraer texto de entrada y texto de etiqueta]
    E --> F[Tokenizar texto de entrada con max_length=512]
    E --> G[Tokenizar texto de etiqueta con max_length=2]
    F --> H[Imprimir tokens de entrada]
    H --> I[Imprimir longitud total de tokens de entrada]
    G --> J[Imprimir tokens de etiqueta]
    J --> K[Imprimir longitud de tokens de etiqueta]
    I --> L[Fin]
    K --> L
    """
    
    print("Código Mermaid para generar el diagrama:")
    print(mermaid_code)
    
    # Guardar el código mermaid en un archivo
    with open("diagrama_flujo_tokenizacion.mmd", "w") as f:
        f.write(mermaid_code)
    
    print("Código Mermaid guardado en 'diagrama_flujo_tokenizacion.mmd'")
    return True

if __name__ == "__main__":
    print("Generando diagrama de flujo del proceso de tokenización...")
    
    # Intentar con graphviz primero
    if not crear_diagrama_flujo_tokenizacion():
        # Si falla, intentar con matplotlib
        if not crear_diagrama_matplotlib():
            # Si todo falla, al menos generar el código mermaid
            generar_codigo_mermaid()
            
```

```python
import graphviz

def crear_diagrama_flujo_entrenamiento_t5():
    """
    Genera un PNG llamado 'diagrama_flujo_entrenamiento_t5.png'
    que describe el flujo:
        1) Cargar modelo T5
        2) Definir compute_metrics
        3) Configurar TrainingArguments
        4) Instanciar Trainer
        5) Ejecutar trainer.train()
    """
    dot = graphviz.Digraph(
        'diagrama_flujo_entrenamiento_t5',
        comment='Proceso de entrenamiento con T5 y HuggingFace',
        format='png'
    )

    # Estilo general
    dot.graph_attr.update(rankdir='TD', size='8,11', dpi='300')
    dot.node_attr.update(shape='box', style='filled',
                         fillcolor='white', fontname='Arial')
    dot.edge_attr.update(fontname='Arial')

    # Nodos
    dot.node('A', 'Inicio', shape='oval')

    dot.node('B', 'Cargar modelo\nT5ForConditionalGeneration\nfrom_pretrained("t5-base")')

    dot.node('C', 'Definir función\ncompute_metrics(eval_pred)\n• accuracy\n• precision/recall/F1')

    dot.node('D', 'Configurar TrainingArguments\n(output_dir, epochs=4,\n batch 16, LR=3e-5, etc.)')

    dot.node('E', 'Instanciar Trainer(\n model, args,\n train_ds_tok,\n test_ds_tok,\n tokenizer,\n compute_metrics)')

    dot.node('F', 'trainer.train()')

    dot.node('G', 'Fin', shape='oval')

    # Conexiones
    dot.edge('A', 'B')
    dot.edge('B', 'C')
    dot.edge('C', 'D')
    dot.edge('D', 'E')
    dot.edge('E', 'F')
    dot.edge('F', 'G')

    # Renderizar y guardar
    dot.render('diagrama_flujo_entrenamiento_t5', cleanup=True)
    print("Diagrama generado como 'diagrama_flujo_entrenamiento_t5.png'")

if __name__ == '__main__':
    crear_diagrama_flujo_entrenamiento_t5()
```
